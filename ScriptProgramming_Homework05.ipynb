```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return '[Count (step: {0})] {1}'.format(self.step, self.count)
    
    def incr(self):
        self.count = self.count+self.step
        return '[Count (step: {0})] {1}'.format(self.step, self.count)
    def __call__(self):
        self.incr()
    
    def __add__(self, y):
        return Counter(self.count + y)
    __radd__ = __add__

    def __sub__(self, other):
        return Counter(self.count - y)
    __rsub__ = __sub__
    
    def __gt__(self, y):
        return self.count > y

    def __ge__(self, y):
        return self.count >= y

    def __lt__(self, y):
        return self.count < y

    def __le__(self, y):
        return self.count <= y

    def __eq__(self, y):
        return self.count == y

    def __ne__(self, y):
        return self.count != y   
    
    
        

c = Counter(10, 1)
print(c)
c.incr()
print(c)
c()
print(c)
print(c==12)
c = c+5
print(c)
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

    - ìš”êµ¬ì‚¬í•­ 2. & ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ êµì§‘í•©ì„ ë°˜í™˜í•œë‹¤.

```python
class MySet(list):
    def __init__(self, l):
        self.l = l
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    #ìš”êµ¬ì‚¬í•­ 1.
    def __or__(self, lst):
        return MySet(set(self) | set(lst))
    
    #ìš”êµ¬ì‚¬í•­ 2.
    def __and__(self, lst):
        return MySet(set(self) & set(lst))
    
    #ìš”êµ¬ì‚¬í•­ 3.
    def __sub__(self, lst):
        return MySet(set(self) - set(lst))
    
    
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])

u = s|t
print(u)

u = s&t
print(u)

s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s-t
print(u)
```
