4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

    - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

\_\_init\_\_() 메소드는 리스트l의 값을 받아서 각 요소를 자기자신에게 추가해준후, 자기자신을 eliminate_duplicate의 인자로 넘겨 중복을 제거한 리스트를 받도록하는 함수이다.  
\_\_str\_\_()메소드는 리스트를 집합 형태로 출력하기 위해서 오버라이딩한 함수이다.  
eliminate_duplicate()는 매개변수로 받은 리스트에 있는 요소들을 중복제거하는 함수이다.

    - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    >     >>> s = MySet([1, 2, 3])
    >     >>> t = MySet([3, 4, 5])
    >     >>> u = s - t
    >     >>> print(u)
    >     MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        self.l = l
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", "
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    #요구사항 1.
    def __or__(self, lst):
        return MySet(set(self) | set(lst))
    
    #요구사항 2.
    def __and__(self, lst):
        return MySet(set(self) & set(lst))
    
    #요구사항 3.
    def __sub__(self, lst):
        return MySet(set(self) - set(lst))
    
    
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])

u = s|t
print(u)

u = s&t
print(u)

s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s-t
print(u)
```

##### 2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

- getWordFrequence() 메소드에 reverse 라는 인자를 만들고 True 또는 False를 인자로 받을 수 있도록 한다.

    - reverse 인자의 디폴트 값은 False 이며, 기본적으로 내림차순으로 정렬한다.

    >     [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

    >     >>> w4.getWordsFrequency(reverse=true)
    >     [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]
